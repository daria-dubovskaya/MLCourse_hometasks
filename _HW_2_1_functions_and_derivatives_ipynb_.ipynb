### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f = 10 * x - 5
df = sp.diff(f, x)
print(f"Похідна функції: {df}")

f_lambdified = sp.lambdify(x, f)

x_vals = np.linspace(0, 10, 100)
f_vals = f_lambdified(x_vals)
df_vals = np.full(x_vals.shape, float(df))

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label="f(x) = 10x - 5", color='blue')
plt.plot(x_vals, df_vals, label="f'(x) = 10", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Фукнкція f(x)=10x−5 є лінійною і показує, що оцінка зростає з росток кількості годин потрачених на вивчення. Похідна функціі постійна і дорівнює 10, що означає, що приріст оцінки постійний і з кожною витраченою годиною зростає на 10.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f = sp.sin(x)
df = sp.diff(f, x)
print(f"Похідна функції: {df}")

x_vals = np.linspace(0, 10, 100)
f_vals = sp.lambdify(x, f)(x_vals)
df_vals = sp.lambdify(x, df)(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label="f(x) = sin(x)", color='blue')
plt.plot(x_vals, df_vals, label="f'(x) = cos(x)", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = sp.symbols('x')
f = sp.ln(x)
df = sp.diff(f, x)
print(f"Похідна функції: {df}")

x_vals = np.linspace(0.1, 10, 100)
f_vals = sp.lambdify(x, f)(x_vals)
df_vals = sp.lambdify(x, df)(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label="f(x) = ln(x)", color='blue')
plt.plot(x_vals, df_vals, label="f'(x) = 1/x", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)
print(f"Похідна функції: {df}")

x_vals = np.linspace(0, 100, 400)
f_vals = sp.lambdify(x, f)(x_vals)
df_vals = sp.lambdify(x, df)(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label="f(x) = x2−5x+1", color='blue')
plt.plot(x_vals, df_vals, label="f'(x) = 2x-5", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
